```python
import pandas as pd
import autosklearn
```

```python
pd = pd.read_csv('../input/warehouse-bins/merged_data.csv')
from sklearn.model_selection import train_test_split
df = pd
train, test = train_test_split(df, train_size=0.90)
y_train = train['3']
x_train = train.drop(['3'], axis='columns')
x_train = x_train.drop(['2'], axis='columns')


y_test = test['3']
x_test = test.drop(['3'], axis='columns')
x_test.drop(columns='2', inplace=True)

# y_train
# print(x_train)
# print(x_test)
```

```python
import autosklearn.classification
```

```python
cls = autosklearn.classification.AutoSklearnClassifier(n_jobs=-1, metric=autosklearn.metrics.accuracy,)
cls.fit(x_train, y_train)

score = cls.score(x_test, y_test)
print(score)

predictions = cls.predict(x_test)
```

```python
import sklearn
print("Accuracy score {:g} using {:s}".
      format(sklearn.metrics.accuracy_score(y_test, predictions),
             cls.automl_._metric.name))
```
