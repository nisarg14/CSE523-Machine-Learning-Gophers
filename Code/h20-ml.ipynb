```python
import h2o
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.automl import H2OAutoML
h2o.init()
import pandas as pd
```

```python
hf = h2o.import_file('../input/warehouse-bins/merged_data.csv', header=1)
hf
```

```python
train, valid = hf.split_frame(ratios = [.95], seed = 1234)
```

```python
predictors = ["0", "100", "101", "102","103","104","105","106","107","108","109","110","111","112","113","114","115","116","117", "118","119"]
response = "3"
```

```python
data = H2OAutoML(stopping_metric = "auc", verbosity="info")
```

```python
data.train(x=predictors, y=response, training_frame=train, validation_frame=valid)
```
